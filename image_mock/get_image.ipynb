```python
from selenium import webdriver
from lxml import etree
from time import sleep
import re
import os
import json
import random
import uuid
```

```python
pattern_w = r'\?w=(\d+)'
pattern_h = r'&h=(\d+)'
```

```python
bro = webdriver.Chrome()

#å¯ä»¥è°ƒæ•´â€œ&word=â€åé¢çš„æœç´ å…³é”®è¯ï¼Œä»¥çˆ¬å–ä¸åŒç±»å‹çš„å›¾ç‰‡åœ°å€
bro.get('https://image.baidu.com/search/index?tn=baiduimage&word=æ‹æ‘„')
sleep(10)

page_text = bro.page_source
tree = etree.HTML(page_text)
item = []

ul_list = tree.xpath('//*[@id="waterfall"]/ul')
i = 1
for ul in ul_list:
    li_list = ul.xpath('./li')
    # print(i)
    for li in li_list:
        try:
            url = li.xpath('./div/div/div[1]/a/div[1]/img/@src')[0]
            title=li.xpath("./div/a/text()")
            wide = re.search(pattern_w, url).group(1)  # search() ä»å­—ç¬¦ä¸²ä»»æ„ä½ç½®æŸ¥æ‰¾åŒ¹é…
            height = re.search(pattern_h, url).group(1)  # search() ä»å­—ç¬¦ä¸²ä»»æ„ä½ç½®æŸ¥æ‰¾åŒ¹é…
            print(wide)
            print(height)
            print(title)
            print(url)
            item.append(
                [
                    title,url,wide,height
                 ]
            )
        except Exception as e:
            continue

    # i+=1
bro.quit()

    
```
